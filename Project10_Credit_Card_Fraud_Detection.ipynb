```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

**DATA COLLECTION AND PROCESSING**

```python
credit_card_data = pd.read_csv('/content/10_creditcard.csv')
```

```python
# First 5 Rows Of The DataFrame
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
# Number Of Rows And Columns
credit_card_data.shape
```

```python
# Information Of Dataset
credit_card_data.info()
```

```python
# Checking The Missing Values
credit_card_data.isnull().sum()
```

```python
# Checking Distribution Of Legit Transactions And Fraudulent Transactions
credit_card_data['Class'].value_counts()
```

,where 0 is the Normal Transaction and 1 is the Fraudulent Transaction.

This Dataset is highly unbalanced.

```python
# Separating The Data For Analysis
legit = credit_card_data[credit_card_data.Class ==0]
fraud = credit_card_data[credit_card_data.Class ==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# Statistical Description Of The Dataset
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# Comparing The Values For Both Transactions
credit_card_data.groupby('Class').mean()
```

**UNDER-SAMPLING**

BUILDING A SAMPLE DATASET CONTAINING SIMILAR DISTRIBUTION OF NORMAL AND FRAUDULENT TRANSACTIONS

NUMBER OF FRAUDULENT TRANSACTIONS : 492

```python
legit_sample = legit.sample(n=492)
```

**CONCATENATING THE TWO DATAFRAMES**

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

**SPLITTING THE DATA INTO FEATURES AND TARGETS**

```python
X = new_dataset.drop('Class', axis=1)
Y = new_dataset['Class']

print(X)
print(Y)
```

**SPLITTING INTO TRAINING AND TEST DATASET**

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

**MODEL TRAINING**

LOGISTIC REGRESSION

```python
model = LogisticRegression()
```

```python
# Training The Logistic Regression Model With Training Data
model.fit(X_train, Y_train)
```

**MODEL EVALUATION**

ACCURACY SCORE

```python
# Accuracy On Training Data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)

print('TRAINING DATA ACCURACY :', training_data_accuracy)
```

```python
# Accuracy On Test Data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)

print('TEST DATA ACCURACY :', test_data_accuracy)
```
